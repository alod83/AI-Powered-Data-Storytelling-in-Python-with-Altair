```python
df2 = pd.read_csv('source/population.csv')
df2.head()
```

```python
df2 = df2[df2['TIME_PERIOD'] == 2021]
```

```python
len(df2)
```

```python
df2.REF_AREA.unique()
```

```python
['Y_UN17', 'Y18-34', 'Y35-54', 'Y_GE55']

group_ages = {'Y_UN17' : [0,18],
              'Y18-34' : [18, 35],
              'Y35-54' : [35,55],
              'Y_GE55' : [55,150]}

data = {}

for k,v in group_ages.items():
    
    ranges = []
    for i in range(v[0],v[1]):
        ranges.append(f"Y{i}")
        
    data[k] = df2[df2['AGE'].isin(ranges)]['AGE'].count()
    
    if k == 'Y_GE55':
        data[k] += df2[df2['AGE'] == 'Y_GE100']['AGE'].count()
        
        
    
data
```

```python
total = 0
for k,v in data.items():
    total += v
    
total
```

```python
len(df2) - len(df2[df2['AGE'] == 'TOTAL'])
```

```python
df3 = pd.DataFrame.from_dict(data,orient='index').reset_index()

df3.rename(columns={'index' : 'AGE', 0 : 'Value'}, inplace=True)
df3
```

```python
df4 = pd.read_csv('source/homeless_by_regions_2021.csv', sep=';')
df4
```

```python
df4 = df4[['ITTER107','Territorio','SEXISTAT1', 'ETA1', 'CITTADINANZA', 'Value']]
```

```python
df4.CITTADINANZA.unique()
```

```python
df4_tot = df4[(df4['ETA1'] == 'TOTAL') & (df4['SEXISTAT1'] == 'T') &  (df4['CITTADINANZA'] == 'TOTAL')]
```

```python
df4_tot['Territorio'].unique()

df4_tot['Territorio'] = df4_tot['Territorio'].str.replace('Trentino Alto Adige / Südtirol', 'Trentino-Alto Adige/Südtirol')
```

```python
# Load the TopoJSON file by URL
url = "https://raw.githubusercontent.com/openpolis/geojson-italy/master/topojson/limits_IT_regions.topo.json"


source = alt.topo_feature(url, "regions")

alt.Chart(source).mark_geoshape().encode(
    tooltip='properties.reg_name:N',
    color=alt.Color('Value:Q',scale=alt.Scale(range=['#075630', '#A4D333','#FCE135',  '#FEB41D','#DB7222']))
).project('mercator').properties(
    width=500,
    height=300
).transform_lookup(
    lookup='properties.reg_name',
    from_=alt.LookupData(df4_tot, 'Territorio', ['Territorio', 'Value'])
).properties(title='Homeless in Italy in 2021')
```

```python
df5_tot = df5[(df5['ETA1'] == 'TOTAL') & (df5['Sesso'] == 'totale')]
df5_tot['Territorio'] = df5_tot['Territorio'].str.replace('Trentino Alto Adige / Südtirol', 'Trentino-Alto Adige/Südtirol')
```

```python
df5_tot = df5_tot[['Value', 'ITTER107']]
df4_tot = df4_tot[['Value', 'ITTER107','Territorio']]
```

```python
df6 = df5_tot.set_index('ITTER107').join(df4_tot.set_index('ITTER107'),lsuffix='_pop', rsuffix='_hom').reset_index()
```

```python
df6['Ratio'] = df6['Value_hom']/df6['Value_pop']
```

```python
# Load the TopoJSON file by URL
url = "https://raw.githubusercontent.com/openpolis/geojson-italy/master/topojson/limits_IT_regions.topo.json"


source = alt.topo_feature(url, "regions")

alt.Chart(source).mark_geoshape().encode(
    tooltip='properties.reg_name:N',
    color=alt.Color('Ratio:Q')
).project('mercator').properties(
    width=500,
    height=500
).transform_lookup(
    lookup='properties.reg_name',
    from_=alt.LookupData(df6, 'Territorio', ['Territorio', 'Ratio'])
).properties(title='Homeless in Italy in 2021')
```
