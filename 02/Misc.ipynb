```python
from config import *
import pandas as pd
import altair as alt

df = pd.DataFrame({
   'X' : [1,2,3],
   'Y' : ['A','B','C']
})

chart = alt.Chart(df).mark_bar(color=iColor
).encode(
    x = 'X:Q',     
    y = 'Y:N' 
).properties(title='A Bar Chart', height=100)

configure_layout(chart)
```

```python
chart = alt.Chart(df).mark_arc(color=iColor
).encode(
    theta = 'X:Q',     
    color = alt.Color('Y:N', scale=alt.Scale(range=color6))
).properties(title='A Pie Chart', height=100)

configure_layout(chart)
```

```python
import random

df = pd.DataFrame({
   'X' : list(range(100)),
   'Y' : [random.randint(0, 100) for _ in range(100)]
})

chart = alt.Chart(df).mark_line(color=iColor
).encode(
    x = 'X:Q',     
    y = 'Y:Q' 
).properties(title='A Line Chart', height=100)

configure_layout(chart)
```

```python
chart = alt.Chart(df).mark_circle(color=iColor
).encode(
    x = 'X:Q',     
    y = 'Y:Q' 
).properties(title='A Scatter Plot', height=100)

configure_layout(chart)
```
